```python
import pandas as pd 
df = pd.read_csv("FED_1867_present.csv")

df = df[df['Election_Type']=="General"]

#df.set_index('Candidate', inplace=True)
incum = pd.read_excel("Incumbents.xlsx")
print("# of incum:",len(incum))
incum = pd.DataFrame.drop_duplicates(incum)
incum.dropna(subset=['Constituency at Next Election'], inplace=True)
print("# of incum:",len(incum))
print(list(incum))
```

```python
print(list(incum))
```

```python
#Link incumency to elections results by unique candidates by name, district and election date, party

incum["uid"] = incum["Name"] + incum["Constituency at Next Election"]+ incum['Political Affiliation']

for i,date in enumerate(list(df["Election_Date"].unique())):
    incum2 = incum[incum['Date of Next Election']== date]
    names = list(incum2["uid"].unique())
    if i == 0:  
        df2 = df[df['Election_Date']== date]
        df2["uid"] = df2["Candidate"] + df2['Constituency']+ df2['Political_Affiliation']
        
        df2['incumbency'] = df2["uid"].apply(lambda x: "incumbent" if x in names else 0)
    else:
        df3 = df[df['Election_Date']== date]
        df3["uid"] = df3["Candidate"]+df3['Constituency']+df3['Political_Affiliation']
        df3['incumbency'] = df3["uid"].apply(lambda x: "incumbent" if x in names else 0)
        df2 = df2.append(df3)


#add markdown of "incumbent" subtext for candidates in table
incums = df2[df2["incumbency"]=="incumbent"]
incums["Candidate"] = incums["Candidate"].apply(lambda x: x +"^"+"Incumbent"+"^")
df2 =  df2[df2["incumbency"]==0]
df2 = df2.append(incums)

#Add markdown link to check mark image for elected candidates
checkmark = """[![image](https://upload.wikimedia.org/wikipedia/commons/thumb/0/03/Green_check.svg/30px-Green_check.svg.png)]
"""
df2['Result'] = df2["Result"].apply(lambda x: checkmark if x == "Elected" else '')

#df2.to_csv("proccessed.csv")

#remove information not included in table
drop_col =['uid','Election_Type', 'incumbency','Last_Name', 'First_Name', 
           'Middle_Names','Occupation','Gender','Parliament','Province_Territory']

df2 =df2.drop(drop_col,axis="columns")
#df2 = df2[['Political_Affiliation','Candidate','Result','Votes',"Election_Date",'Constituency']]
#df2.columns = ['Party','Candidate','Elected','Votes',"Election_Date",'Constituency']
df2.to_csv("LOP2_linked.csv")


ele = list(df2["Election_Date"].unique())

```

```python
def dataW_chart(title,df,template=["tables", "9dUYf"]):

    if title in list(district_tables.keys()):

        x = district_tables[title]  # [:-46] + """width="2000" height="900"></iframe>"""
        return x
    else:
        chart_id, district_tables[title] = create_dw_chart(df, template,
                                                           file_path=None,
                                                           title=title, )

        x = district_tables[title]  # [:-46] + """width="2000" height="900"></iframe>"""
        # Save chart for later :)
        with open('district_table_iframes.pickle', 'wb') as handle:
            pickle.dump(district_tables, handle, protocol=pickle.HIGHEST_PROTOCOL)

        return x

def party_table_data(election):
    # load and filter candidate data
    file2 = 'LOP2_linked.csv'
    full_table = pd.read_csv(file2)
    party_df = full_table[full_table["Election_Date"] == election]
    # Only look at elected candidates
    c = list(party_df["Result"].unique())
    party_df = party_df[party_df["Result"] == c[1]]
    district_summary = party_df.groupby(['Constituency', 'Political_Affiliation'], as_index=False)[
        'Votes'].sum()
    district_summary = district_summary[['Constituency', 'Political_Affiliation']]
    district_summary.columns = ['Constituency', 'Elected Candidate Affiliation']

    return district_summary

def provincial_summary(df,election):
    df_election = df[df["date"] == election]

    prov_summary = df_election.groupby(['province'], as_index=False)['vote_total'].sum()
    prov_summary = prov_summary[['province', 'vote_total']]
    prov_summary.columns = ['Province', 'Total Valid Votes']
    return df_election,prov_summary

def create_dw_chart(data, template, file_path=None, title=" ", ):
    """
    Returns iframe, png of datawrapper chart based on template and data

    """
    # create Datawrapper chart
    chart = dw.create_chart(title=title, chart_type=template[0], data=data, folder_id="District_tables")

    # get meta data of chart template
    chart_template = dw.chart_properties(template[1])
    dw.update_metadata(chart['id'], chart_template["metadata"])

    # publish chart and get iFrame code to embed in Streamlit
    dw.publish_chart(chart_id=chart['id'])

    if file_path:
        dw.export_chart(chart['id'], output='png', filepath=file_path, width=1000,
                        border_width=20, plain=True, mode="cmyk")

    return chart['id'], dw.get_iframe_code(chart['id'])

# folder = "/Users/dylanclark/Desktop/district_summaries.csv"
# df = pd.read_csv(folder)
# election_dates = df["date"].unique()



# file ='/Users/dylanclark/Desktop/LOP_linked.csv'
# candidate_table = pd.read_csv(file)

def incumbent_table():
    """
    Function to create incumbent tables for all of the federal elections
    """
    incum = pd.read_excel('Incumbents.xlsx')
    incum = pd.DataFrame.drop_duplicates(incum)
    incum.dropna(subset=['Constituency at Next Election'], inplace=True)
    for election in list(incum["Date of Next Election"].unique()):
        incum2 = incum[incum["Date of Next Election"] == election]
        incum2 = incum2[['Name', 'Constituency at Next Election', 'Province/Territory']]
        incum2.columns = ['Name', 'Constituency', 'Province/Territory']
        title = election + ", Incumbent Elects Seeking Re-Election"
        district_tables[title] = dataW_chart(title,incum2,template=["tables", "cBFId"])
        
        with open("/Users/dylanclark/Desktop/" + 'district_table_iframes.pickle', 'wb') as handle:
            pickle.dump(district_tables, handle, protocol=pickle.HIGHEST_PROTOCOL)


    return incum

#incumbent_table()
```

```python
from tqdm import tqdm
import pandas as pd
import pickle

folder = "district_summaries.csv"
df = pd.read_csv(folder)
with open('data_base_parms.pickle', 'rb') as handle:
    data_base_parms = pickle.load(handle)
    
for election in tqdm(data_base_parms["elections"]):
    df_election, prov_summary = provincial_summary(df, election)
    # Unique districts
    ridings = list(df_election["district"].unique())
    ridings.append("Summary")
    data_base_parms[election + " ridings"] = ridings
    
with open('data_base_parms.pickle', 'wb') as handle:
    pickle.dump(data_base_parms, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
